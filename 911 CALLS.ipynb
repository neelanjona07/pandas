```python
df['month']=df['timeStamp'].apply(lambda l:l.month)
```

```python
df['day']=df['timeStamp'].apply(lambda l:l.day)
```

```python
df['hour']=df['timeStamp'].apply(lambda l:l.hour)
```

```python
df['dayofweek']=df['timeStamp'].apply(lambda l:l.dayofweek)
```

```python
df['date']=df['timeStamp'].apply(lambda l:l.date())
```

```python
df
```

```python
d={0:'mon',1:'tue',2:'wed',3:'thurs',4:'fri',5:'sat',6:'sun'}
```

```python
df['dayofweek']=df['dayofweek'].map(d)
```

```python
df
```

```python
sns.countplot(x='dayofweek',data=df,hue='Reason')
```

```python
sns.countplot(x='month',data=df,hue='Reason')
```

```python
df.groupby('date').count()['twp'].plot()
```

```python
df.groupby('month').count()['twp'].plot()
```

```python
df[df['Reason']=='Fire'].groupby('date').count()['twp'].plot()
plt.title('Fire')
plt.tight_layout()
```

```python
df[df['Reason']=='Traffic'].groupby('date').count()['twp'].plot()
plt.title('Traffic')
plt.tight_layout()
```

```python
df[df['Reason']=='EMS'].groupby('date').count()['twp'].plot()
plt.title('EMS')
plt.tight_layout()
```
